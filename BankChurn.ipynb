```python
#hago el grid search
#params de DecTreeClass
parameter_grid_dtc= {
    "max_depth": [5,10,20,25,50,80],
    "max_features": [0.3, 0.5, 1],
    "min_samples_split": [5,20,50,100,150, 300]
}
grid_search_dtc= GridSearchCV(DecisionTreeClassifier(random_state=0), parameter_grid_dtc, cv=3, verbose=3, refit=True, n_jobs=-1, scoring="neg_mean_squared_error") 
```

```python
dmatrix = xgb.DMatrix(data=X, label=y)
```

```python
params={'objective':'reg:squarederror'}
cv_results = xgb.cv(dtrain=dmatrix, params=params, nfold=10, metrics={'rmse'}, as_pandas=True, seed=20)
```

```python
#con hiperparametros
params={ 'objective':'reg:squarederror',
         'max_depth': 6, 
         'colsample_bylevel':0.5,
         'learning_rate':0.01,
         'random_state':20}
cv_results = xgb.cv(dtrain=dmatrix, params=params, nfold=10, metrics={'rmse'}, as_pandas=True, seed=20, num_boost_round=100)
print('RMSE: %.2f' % cv_results['test-rmse-mean'].min())
```
